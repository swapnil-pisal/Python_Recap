# Functions & Modules

```python
# Functions 
# user defined function  
# def keyword use the create the function 
def calculater(a,b):
    mean=(a+b)/2
    print(f"The Avarage is {mean} ")

calculater(100,10)
calculater(23,134)
calculater(23,23)
calculater(45,78)
calculater(65,12)
```

```python
def isgreater(x,y):
    if x>y :
        print("X is greater ")
    else: 
        print("Y is greater")

isgreater(34,45)
isgreater(56,23)
isgreater(23,67)
```

```python
# Docstring

def show(n): 
    ''' This is Docstring line it is the first line of the method
     '''
    print(n**2)

show(5)
print(show.__doc__)
```

```python
#Find the factorial number 
# 5=5*4*3*2*1

def factorial(n):
    if (n==0 or n==1):
        return 1 
    else:
        return n*factorial(n-1)

print(factorial(5))
print(factorial(3))
print(factorial(7))
print(factorial(9))
print(factorial(3))
print(factorial(10))
    
```

```python
# enumerate function 
# it is use for direct index 

marks=[56,67,78,89,98,99]

for index,mark in enumerate(marks,start=1):
    print(mark)
    if index == 4: 
        print("Good")
```

```python
# lambda function 
# it is anamonuname functions ( name less function )

square = lambda x:x*x 

cube = lambda y:y*y*y             # they can be use the multiple variable

additon = lambda a,b:a+b  

div = lambda p,q:(p+q)/2

print(square(5))

print(cube(5))

print(additon(23,45))

print(div(6,4))
```

```python
# map function 

def cube(n):
    return n*n*n
print(cube(12))

list1=[12,23,34,45]
print(list1)

# for i in list1:
#     newl=[] 
#     newl.append(cube(i))
#     print(newl)

# skip the above three line code use map function

newl=list(map(cube,list1))
print(newl)

# map function = combine the function and list
```

```python
# filter function 

list2=[98,98,97,76,4,90]
print(list2)

def func_filter(a):
    return a>=90
    
    
n=list(filter(func_filter,list2))
print(n)
```
