```python
# print different way
num=32
print("The number is : ",num)
print("The type of number is  : ",type(num))

# newline charcter
print("The number is : \n",num)
print("The type of number is  : \n",type(num))

# f string
print(f"The number is : {num}")
print(f"The type of number is  : \n",type(num))
```
