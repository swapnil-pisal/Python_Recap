```python
# numbers list
list1=[1,2,3,4,5,6,7]
print(list1)
```

```python
# String list
list2=["Swapnil","Rohan","Ram","Sham"]
print(list2)
```

```python
# floating point list
list3=[12.22,15.33,55.44]
print(list3)
```

```python
# boolean list
list4=[True,False,0,1]
print(list4)
```

```python
# mixed list
list5=[1,"Mango","Banana",-23,-33,True,34]
print(list5)
```

```python
list6=["Mumbai","Pune","Delhi","Bengluru","Kolkata","Chenai"]
print(list6)
```

# list method

```python
print(list6)
```

```python
# Add an item to the end of the list.
list6.append("Ranchi")
print(list6)
```

```python
# Extend the list by appending all the items . 
list6.extend(["Goa","Patana","Laknow"])
print(list6)
```

```python
# Insert an item at a given position. 
list6.insert(3,"Punjab")
print(list6)
```

```python
# remove the element from the list
list6.remove("Patana")
print(list6)
```

```python
list6.pop()
print(list6)
```

```python
list6.extend(["Mumbai","Mumbai","Mumbai"])
print(list6)
```

```python
# Return the number of times Mumbai appears in the list. 
print(list6.count("Mumbai"))
```

```python
# sort the list
i=sorted(list6)
print(i)
```

```python
# Reverse the elements of the list in place
r=list6.reverse()
print(r)
```

```python
# delete statement 
del list6[11:13]
print(list6)
```

```python
# delete the specific element 
del list6[6]
print(list6)
```

```python
list8=[23,34,45,66,12,55,66,75]
print(list8)
```

```python
# accesnding order 
l2=sorted(list8)
print(l2)
```

```python
# decending order 
l3=sorted(list8,reverse=True)
print(l3)
```

```python
# lowercase all element in the list  
print(list6) 
lw=[l.lower() for l in list6]
print(lw)
```

```python
# update the element inn list
print(list6)
list6[9]="Hydrabad"
print(list6)
```
