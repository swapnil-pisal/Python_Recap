# Dictionaries

It is set of key: value pairs

It is in the curly braces {}

```python
dict1 = {
    "Name" : "Swapnil" , 
    "Age" : 21 , 
    "Occ" : "Data Science" , 
    "Phone Number" : 9370172934    
}
```

```python
print(dict1)
```

```python
dict2={
    1 : "Ram" ,
    2 : "Sham" ,
    3 : "Seeta" , 
    4 : "Geeta" ,
    5 : "Mihir"
}
```

```python
print(dict2)
```

```python
dict3={
    "Name" : ["Swapnil","Vijay","Vivek","Avinash"], 
    "Marks" : [98,99,97,90], 
    "Subject" : ["Maths","Database","Java","Software"]
}
```

```python
print(dict3)
```

```python
dict4={
    "Name" : ("Seeta","Govind","Vivek","Ram"), 
    "Marks" : (98,99,97,90), 
    "Subject" : ("Maths","Database","Java","Software")
}
```

```python
print(dict4)
```

```python
print(dict1)
```

```python
# add the elemnt in the dictinory
dict1["CGPA"]=8.5
```

```python
print(dict1)
```

```python
# access using keys 
print(dict1["Name"])
```

```python
# Access using get method
print(dict1.get('Age'))
```

```python
# delete the element in dictionary 
del dict1["Phone Number"]
print(dict1)
```

```python
if "Name" in dict1:
    print("yes")
else:
    print("No")
```

```python
for i in dict1:
    print(i)
```

```python
print(dict1.keys())
```

```python
print(dict1.values())
```

```python
# Both key and value use items() method 
print(dict1.items())
```

```python
for key,value in dict1.items():
     print(f"the student {key} is {value}")
```
