# Tuples

Tuple are immutable.

while list are mutable.

List can be chage after creation but tuple can not change after creationn

```python
# tuples 
# interger
tuple1=(12,11,43,24,55,45)
print(tuple1)
```

```python
# String
tuple2=("Swapnil","Vijay","Vivek","Avinash","NIkhil")
print(tuple2)
```

```python
# Negative number 
tuple3=(-34,-45,-46,23)
print(tuple3)
```

```python
# mixed values
tuple4=(12,34,"Mumbai","Pune",23.22,34.34,True,False,-34,-87)
print(tuple4)
```

```python
tuple5=(122,23,3344,556,67,(23,34,4556,45),34.45,56,67,(34,56,67,78))
print(tuple5)
```

```python
tuple6=("Mango","Banana","Graps")
print(tuple6)
```

```python
# access the tuple with index
print(tuple6[0])
```

```python
print(tuple6[0:2])
```

```python
print(tuple6[-1])
```
