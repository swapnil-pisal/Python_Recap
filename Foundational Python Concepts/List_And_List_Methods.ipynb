```python
#sorting the list in ascending using the sorted()

l = [7,8,9,6,1,0,5,100,300,5000,-5,-10,-1] #list of different elements

asc = sorted(l)
print(asc)
```

```python
#sorting the list in descending using the sorted() and reverse=True

l = [7,8,9,6,1,0,5,100,300,5000,-5,-10,-1] #list of different elements

dec = sorted(l, reverse=True)
print(dec)
```

```python
l1 = ["b","A","B","R","c","d","h","H","j","r"]

asce = sorted(l1)
loasc = [lst.lower() for lst in asce]
print(loasc)
```

```python
#Adding a element or elements in the list using append() and extend([])

l2 = [6,7,9,0]

#I want to add new element in the given list
l2.append(75)

# More elment add in the list
l2.extend([99,100,101,-50])

a1 = sorted(l2)
print(a1)
```

```python
# delete the element from the list using del

lst=[100,200,500,700,40,80]

# index
#0-100,1-200,2-500,3-700,......

del(lst[4])
print(lst)
```

```python
# create a list of strings (6),eg: cars name/brands and you have to delete 3rd element.
cars=["BMW","Mercidies","Safari","RollsRoyal","Ferrari","Lamborghini"]
print(cars)
del(cars[2])
print(cars)
```

```python
# create a list mango, apple, swift and replace swift to a fruit.

fruits = ["mango", "banana", "swift"]

del(fruits[2])
fruits.append("apple")
print(fruits)
```
