 # Control Flow: 
Using if, elif, and else for decision-making, and for and while loops to iterate over datasets.

```python
# Using if, elif, and else for decision-making 

# checking eligibile for voting

age = int (input("Enter the age : "))
if age < 18: 
    print("Sorry ! you not eligible for voting")
else : 
    print("Yes ! you are eligible for voting")
```

```python
# checking number is even or odd

num = int (input("Enter the number : "))
if num % 2 == 0 : 
    print(f"The {num} is Even Number")
else : 
    print(f"The {num} is Odd Number")
```

```python
# the number is positive , neative or zero check  

num=int(input("Enter the number : "))

if num == 0 :
    print(f"{num} it is zero number.")
if num <= 0 :
    print(f"{num} it is negative number.") 
else :
    print(f"{num} it is positive number.")
```

```python
# Student mark analysis using if elif else 

marks=int(input("Enter the marks : "))

if marks>0 and marks<=100:
    if marks <= 100 and marks >= 85 :
        print(f"You got the {marks} with A grade.")
    elif marks < 85 and marks >= 60 :
        print(f"You got the {marks} with B grade.")
    elif marks < 60 and marks >= 35 :
        print(f"You got the {marks} with C grade.")
    else : 
        print(f"You fail.")
else : 
    print("Invalid Marks")
        
```

```python
# for loop 

# print the tables of Eg.6

num=int(input("Enter the number : "))

for i in range (1,11):
    print(f"{num} X {i} = {num*i}")
```

```python
#Palindrome eg: nitin == nitin

s3 = input()

if s3 == s3[::-1]:
  print("The given string is palindrome.")
else:
  print("The given string is not palindrome.")
```

```python
#star square pattern

n = int(input("Enter a number: "))

if n > 0:
  for i in range(1,n+1):
    print("*"*n)
else:
  print("please enter number greater than 0.")
```

```python
#star triangle pattern

n = int(input("Enter a number: "))

if n > 0:
  for i in range(1,n+1):
    print("*"*i)
else:
  print("please enter number greater than 0.")
```

```python
#hash reverse triangle pattern

n = int(input("Enter a number: "))

if n > 0:
  for i in range(n,0,-1):
    print("#"*i)
else:
  print("please enter number greater than 0.")
```

```python
#pyramid pattern

n = int(input("Enter a number: "))

if n > 0:
  for i in range(1,n+1):
    print(" " * (n-i),end="")
    print("*" * (2*i-1),end="")
    print()
else:
  print("please enter number greater than 0.")
```

```python
# While loop 

num1=int(input("Enter the first number : "))
num2=int(input("Enter the second number : "))

while num1+num2 != 100: 
    print("Try Again ! ")
    num1=int(input("Enter the first number : "))
    num2=int(input("Enter the second number : "))
print("The Sum Of Num1 And Num2 is 100")
```
