# Sets

Sets are unordered coolection of element

Sets Remove Duplicates

Sets are mutable it mean it can be change after creation we can add or drop element from the sets

Sets Witin Write in Curly braces or set() function 

Empty Sets Are Use set() . Not Use {} because it has the Empty Dictionary

Sets perform Mathematical oeration like union , intersection , set difference , Difference

```python
#using curly braces
set1 = {1,2,3,1,2,3,4,4,5,5,6,2,4}
print(set1)
```

```python
# set function 
a = set([1,2,3,1,2,3,4,4,5,5,6,2,4])
print(a)
```

```python
d = {} 
print(d) 
print(type(d))
```

```python
# empty set 
b = set() 
print(b)
print(type(b))
```

```python
fruits = {"Mango" , "Banana" , "Graps" , "Gava" , "Mango" , "Banana"}
print(fruits)
```

```python
for fruit in fruits: 
    print(fruit)
```

```python
# Add the element in the set
fruits.add("Apple")
print(fruits)
```

```python
# delete the element in sets  
fruits.remove("Gava")
print(fruits)
```

```python
# delete also discard method . if element not in set we can delete error not be found with help of discard method
fruits.discard("Cocnut")
print(fruits)
```

```python
# delete all set use clear method 
fruits.clear()
print(fruits)
```

```python
# Sets perform Mathematical oeration like union , intersection , set difference , Difference
```

```python
# Two set for operation
set1 = {1,2,3,4,5}
set2 = {1,2,5}
print(set1)
print(set2)

# union 
print("Combine the two set and remove duplicate : ",set1 | set2)

# Intersection 
print("Comman between two sets : ",set1 & set2)

# difference 
print("The element is in set1 but not in set2 : ",set1 - set2)

# symmetric difference 
print("The Comman element are remove both sets : ",set1 ^ set2)
```
