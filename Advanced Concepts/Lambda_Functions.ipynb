# Lambda Functions

 Small, anonymous functions useful for quick data transformations

```python
def square (x):
    return x**2
square(2)
```

```python
square = lambda x:x**2

print(square(2))
```

```python
cube = lambda x:x**3

print(cube(3))
```

```python
cube = lambda x:x*x*x

print(cube(3))
```

```python
# it can be use multiple variable in lambda function 

area = lambda x,y:x*y

print(area(4,5))
```

```python
a = lambda x,y:x*y/2

print(a(10,5))
```
