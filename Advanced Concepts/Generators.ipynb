# Generators

Crucial for memory-efficient looping over massive datasets that don't fit in RAM.

Generators can be generate the fly values 

It can be generate the as requred values. they are lazy.

```python
def genrater_function(): 
    for i in range(5): 
        yield i

gen = genrater_function()

print("First Element is : ",next(gen))
print("Second Element is : ",next(gen))
print("Third Element is : ",next(gen))

for j in gen:
    print(j)

list1 = [x for x in range(5)]
print(list1)

import sys
print("The memory used by generators are :",sys.getsizeof(j))
print("The memory used by list are :",sys.getsizeof(list1))
```

```python
# explicitly write a generator function
def double(L): 
    for i in L: 
        yield i*2

# eggs will be a generator
eggs = double([1,2,3,4,5])

print(next(eggs))
print(next(eggs))
print(next(eggs))
```

```python
# the above is equivalent to ("generator comprehension"?)
eggs = (x*2 for x in [1, 2, 3, 4, 5])

print(next(eggs))
print(next(eggs))
print(next(eggs))
```

```python
# if we want a list 

eggs = list(double([1,2,3,4,5]))

print(eggs)
```
