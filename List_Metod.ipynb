```python
#insert element in the list at specific index using insert()
lst = [1,2,3,4,5,6,7]

lst.insert(5, "Audi")
print(lst)
```

```python
#Replace the particular element in the list using slicing

lst[6] = "Maruti"
print(lst)
```

```python
# python operators 'sum','difference','product','dividion',reminder'.

a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))

add = a+b
diff = a-b
pro = a*b
div = round(a/b,2)
sqr = a**2
rem = a%b
fd = a//b

print(f"Sum of {a} and {b} is: {add}.")
print(f"The difference of {a} and {b} is: {diff}.")
print(f"The product of {a} and {b} is: {pro}.")
print(f"The division of {a} and {b} is: {div}.")
print(f"The square of {a} is: {sqr}.")
print(f"The mod of a and b is: {rem}.")
print(f"The fd of a and b is: {fd}.")
```
